## Exercise: Sampling a global seismic wavefield

Slide 17
- The highest frequencies that we observe for globalwave fields is 1Hz.
- We assume a homogeneous Earth (radius 6371km).
- P velocity vp=10km/s and the vp/vs ratio is sqrt(3)
- We want to use 20 grid points (cells) per wavelength
- How many grid cells would you need (assume cubic cells).
- What would be their size?
- How much memory would you need to store one suchfield (e.g., density in single precision)

```python
import numpy as np
```

```python
f = 1  #Hz
T = 1/f  #s
r = 6371*(10**3)  #m
vp = 10*(10**3)  #m/s
vs = vp/np.sqrt(3)  #m/s
#precision: choose single = 4 and double = 8
p = 4
```

```python
#wavelength wl
wl = vs*T
```

```python
#grid points
gp = wl/20
```

```python
#volume of the Earth
V = 4/3 * np.pi * r**3
```

```python
#computational cell size in 3D
dV = gp**3
```

```python
#number of cells nc
nc = V/dV
print(nc)
```

```python
#memory requirement (density in single precision)
mem = p * nc
print(mem)
```

```python
#memory requirement in GB and TB
mem_GB = mem*(10**(-9))
mem_TB = mem*(10**(-12))
print('The memory requirement would {} GB or {} TB'.format(mem_GB, mem_TB))
```

### Plot as funtion of frequency (or period)

```python
import matplotlib.pyplot as plt
```

```python
f = np.arange(0.1, 1.1, 0.1)  #Hz
T = 1/f  #s
r = 6371*(10**3)  #m
vp = 10*(10**3)  #m/s
vs = vp/np.sqrt(3)  #m/s
wl = vs*T
gp = wl/20
#computational ell size in 3D
dV = gp**3
#volume of the Earth V
V = 4/3 * np.pi * r**3
nc = V/dV
mem = p * nc
mem_GB = mem*(10**(-9))
mem_TB = mem*(10**(-12))
print(mem_GB)
```

```python
plt.plot(f,mem_TB)
plt.xlabel('frequency [Hz]')
plt.ylabel('memory single precision [TB] ')
```

### Plot as funtion of frequency (or period) using subplots

```python
f = np.arange(0.1, 1, 0.001)  #Hz
T = 1/f  #s
r = 6371*(10**3)  #m
vp = 10*(10**3)  #m/s
vs = vp/np.sqrt(3)  #m/s
wl = vs*T
gp = wl/20
#computational ell size in 3D
dV = gp**3
#volume of the Earth V
V = 4/3 * np.pi * r**3
nc = V/dV
mem = p * nc
mem_GB = mem*(10**(-9))
mem_TB = mem*(10**(-12))
```

```python
plt.figure(figsize=(20,10))
plt.subplot()

plt.subplot(1,2,1)
plt.plot(f,mem_GB)
plt.xlabel('frequency [Hz]')
plt.ylabel('memory single precision [GB] ')

plt.subplot(1,2,2)
plt.plot(f,mem_TB)
plt.xlabel('frequency [Hz]')
plt.ylabel('memory single precision [TB] ')

plt.suptitle("Memory requirements")

plt.savefig('/import/freenas-m-04-students/fkutschera/Master/Computational_Seismology/Exercises_class/memory_requirements.png', dpi=300)
```
